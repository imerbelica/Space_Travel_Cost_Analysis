```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from scipy import stats

path = r'C:\Users\Imer Belica\Desktop\archive (1)\Space_Corrected.csv'
names = ['','Unnamed','Company Name','Location','Datum','Detail','Status Rocket', 'Rocket','Status Mission','Private or State Run']
dataframe = pd.read_csv(path, header = 0,names=names)
dataframe.dropna(
    axis=0,
    how='any',
    thresh=None,
    subset=None,
    inplace=True
)
dataframe.reset_index(drop=True, inplace=True)

dataframe
```

```python
a = np.array(dataframe.Rocket)
z = stats.zscore(a)
dataframe['Z-Score'] = z
normalData = []
count = 0
for score in z:
    if np.absolute(score)>3:
        normalData.append(count)
    count += 1
dataframe = dataframe.drop(normalData)
dataframe.reset_index(drop=True, inplace=True)
dataframe
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
dt = []
for z in dataframe.Datum:
    if ':' in z:
        z = z[0:-6]
    datetime_object = datetime.strptime( z, '%a %b %d, %Y')
    dt.append(datetime_object)
count = 0
avg = []
current = 2020
Y = []
X = [current]
for cost in dataframe.Rocket:
    if dt[count].year == current:
        avg.append(cost)
        count += 1
    else:
        print(current, ':', len(avg))
        current = dt[count].year
        Y.append(sum(avg) / len(avg))
        avg = [cost]
        X.append(current)
        count += 1
Y.append(sum(avg) / len(avg))

plt.title('Cost Vs Time')
plt.scatter(X,Y)
```

```python
count = 0
avgP = []
avgS = []
current = 2020
Y = []
X = []
color = []
for cost in dataframe.Rocket:
    if dt[count].year == current and dataframe['Private or State Run'][count] == 'S':
        avgS.append(cost)
        count += 1
    elif dt[count].year == current and dataframe['Private or State Run'][count] == 'P':
        avgP.append(cost)
        count += 1
    else:
        if len(avgS) >0:
            Y.append(sum(avgS) / len(avgS))
            X.append(current)
            color.append('red')
        if len(avgP) > 0:
            Y.append(sum(avgP) / len(avgP))
            X.append(current)
            color.append('blue')
        current = dt[count].year
        avgS = []
        avgP = []
        if dataframe['Private or State Run'][count] == 'S':
            avgS.append(cost)
        else:
            avgP.append(cost)
        count += 1       
if len(avgS) >0:
    Y.append(sum(avgS) / len(avgS))
    X.append(current)
    color.append('red')
if len(avgP) > 0:
    Y.append(sum(avgP) / len(avgP))
    X.append(current)
    color.append('blue')


plt.title('Private Vs State Mission Cost')
plt.scatter(X,Y,
            c=color)
```

```python
dfs = pd.DataFrame(columns = ['Date' , 'Cost', 'Status Mission'])
dfs['Date'] = dt
dfs['Cost'] = dataframe.Rocket
dfs['Status Mission'] = dataframe['Status Mission']
dfs.head()
```

```python
import seaborn as sns
favg = []
savg = []
count = 0
for c in dfs.Cost:
    if dfs['Status Mission'][count] == 'Success':
        savg.append(c)
    else:
        favg.append(c)
    count +=1
successAvg = (sum(savg) / len(savg))
failAvg = (sum(favg)/ len(favg))

plt.figure(figsize=(8, 6))
splot=sns.barplot(x="Status Mission",y="Cost",data=dfs)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', color = 'white', 
                   xytext = (0, -100), 
                   textcoords = 'offset points')

plt.xlabel("Status", size=14)
plt.ylabel("Cost", size=14)
plt.title('Cost Vs Mission Status')
plt.show()
```
